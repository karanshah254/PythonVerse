```python
import matplotlib.pyplot as plt

x = ["Python", "C", "C++", "Java"]
y = [85, 50, 60, 90]
z = [20, 30, 40, 50]
colors = ["red", "green", "blue", "yellow"]
plt.xlabel("Languages", fontsize=15)
plt.ylabel("Popularity", fontsize=15)
plt.title("Popularity of Programming Languages", fontsize=20)
plt.bar(
    x,
    y,
    width=0.4,
    color="r",
    edgecolor="black",
    label="Popularity",
)
plt.bar(
    x,
    z,
    width=0.4,
    color="y",
    edgecolor="black",
    label="Popularity",
)
plt.legend(loc="upper center")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["Python", "C", "C++", "Java"]
y = [85, 50, 60, 90]
z = [20, 30, 40, 50]
p = np.arange(len(x))
width = 0.3
p1 = [i + width for i in p]

plt.xlabel("Languages", fontsize=15)
plt.ylabel("Popularity", fontsize=15)
plt.title("Popularity of Programming Languages", fontsize=20)

plt.bar(
    p,
    y,
    width,
    color="r",
    edgecolor="black",
    label="Popularity",
)
plt.bar(
    p1,
    z,
    width,
    color="y",
    edgecolor="black",
    label="Popularity",
)

plt.xticks(p + width/2, x, rotation=20)
plt.legend(loc="upper center")
plt.show()
```
