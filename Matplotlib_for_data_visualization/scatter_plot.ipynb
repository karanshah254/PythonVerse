```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6, 7]
y = [2, 3, 1, 4, 7, 6, 8]
c = ["red", "green", "blue", "yellow", "black", "purple", "orange"]
plt.scatter(x, y, c=c, s=200, marker="*", edgecolors="black")
plt.title("Scatter plot", fontsize=20)
plt.xlabel("x", fontsize=15)
plt.ylabel("y", fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6, 7]
y = [2, 3, 1, 4, 7, 6, 8]
z = [6, 2, 3, 5, 1, 6, 7]
colors = [10, 20, 30, 40, 50, 60, 70]
plt.scatter(x, y, c=colors, s=200, marker="*", edgecolors="black", cmap="cool")
plt.scatter(x, z, c="y", s=200, edgecolors="black", alpha=0.3)
t = plt.colorbar()
t.set_label("color bar", fontsize=15)
plt.title("Scatter plot", fontsize=20)
plt.xlabel("x", fontsize=15)
plt.ylabel("y", fontsize=15)
plt.show()
```
