```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Genrating data
x = np.linspace(0, 10, 100)  # 100 evenly spaced numbers from 0 to 10
y = np.sin(x)  # sine of x
```

```python
# Plotting data
plt.plot(x, y)
plt.title('Sine wave function')
plt.xlabel('Input values')
plt.ylabel('Sine values')
plt.grid(True)
plt.show()
```
