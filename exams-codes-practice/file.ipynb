#### Write a program using Numpy to count number of “C” element wise in a given array.

```python
import numpy as np

arr = np.array(["A", "B", "C", "C", "C", "D", "E", "C"])

no_of_c = np.count_nonzero(arr == "C")
no_of_c
```

#### Write a simple python program that draws a line graph where x = [1,2,3,4] and y = [1,4,9,16] and gives both axis label as “X-axis”and “Y-axis”.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot(x, y)
```

#### Write a Python programming to create a pie chart with a title of the popularity of programming Languages. 
#### Programming languages: Java, Python, PHP, JavaScript, C#, C++ 
#### Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python
import matplotlib.pyplot as plt

languages = ["Java", "Python", "PHP", "JavaScript", "C#", "C++"]
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

plt.title("Popularity of Programming Language")
plt.pie(popularity, labels=languages)
```

#### Check Prime number or not

```python
import math


def isPrime(number):
    if number <= 1:
        return False
    for i in range(2, int(math.sqrt(number)) + 1):
        if number % i == 0:
            return False
    return True


number = int(input("enter number"))
print(f"Is {number} prime? {isPrime(number)}")
```

#### Write a program to print Fibonacci series up to number given by user.

```python
terms = int(input("Enter number of terms: "))
count = 0
n1, n2 = 0, 1
print(f"Fibonacci sequence for {terms} terms is :", end=" ")
while count <= terms:
    print(n1, end=" ")
    nth = n1 + n2
    n1 = n2
    n2 = nth
    count += 1
```
