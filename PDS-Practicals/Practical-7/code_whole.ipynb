```python
import matplotlib.pyplot as plt
import numpy as np
import random

# Part A
X = np.arange(1, 51, 1)
Y = np.multiply(X, 3)
plt.plot(X, Y, "b+")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Plot of X-values vs Y-values")
plt.show()


# Part B
languages = ["Java", "Python", "PHP", "JavaScript", "C#", "C++"]
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
plt.bar(languages, popularity)
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.title("Popularity of Programming Languages")
plt.show()


# Part C
languages = ["Java", "Python", "PHP", "JavaScript", "C#", "C++"]
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
colors = ["red", "blue", "green", "yellow", "purple", "orange"]
plt.pie(popularity, labels=languages, colors=colors, autopct="%1.1f%%")
plt.title("Popularity of Programming Languages")
plt.legend(languages, loc="best")
plt.show()


# Part D
X = [random.randint(1, 201) for x in range(200)]
Y = [random.randint(1, 201) for x in range(200)]
plt.scatter(X, Y)
plt.xlabel("Random X-values")
plt.ylabel("Random Y-values")
plt.xlim(0, 200)
plt.ylim(0, 200)
plt.title("Random X-values vs Random Y-values")
plt.show()
```
