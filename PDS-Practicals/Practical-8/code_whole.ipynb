```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Read the data
sales_data = pd.read_excel(
    "https://github.com/chris1610/pbpython/blob/master/data/sample-salesv3.xlsx?raw=True",
)

# Preprocess the data
sales_data["date"] = pd.to_datetime(sales_data["date"])

# visualize the data
plt.figure(figsize=(10, 6))
plt.plot(sales_data["date"], sales_data["ext price"])
plt.title("Sales data")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()

# save the plot
plt.savefig("sales_data.png")
```
