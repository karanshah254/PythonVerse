```python
import matplotlib.pyplot as plt
%matplotlib inline
values = [5,8,9,4,1,6,7,2,3,8]
plt.plot(range(1,11),values)
plt.show()
```

```python
# draw multiple lines
values = [5, 8, 9, 4, 1, 6, 7, 2, 3, 8]
values2 = [3, 7, 2, 5, 8, 2, 5, 6, 9, 12]
plt.plot(range(1, 11), values)
plt.plot(range(1, 11), values2)
plt.show()
```

```python
# axis, grid, ticks
values = [5, 8, 9, 4, 1, 6, 7, 2, 3, 8]
ax = plt.axes()
ax.set_xlim([0, 50])
ax.set_ylim([-10, 10])
ax.set_xticks([0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50])
ax.set_yticks([-10, -8, -6, -4, -2, 0, 2, 4, 6, 8, 10])
ax.grid()
plt.plot(range(1, 11), values)
plt.show()
```

# Line appearance

```python
values1 = [5, 8, 9, 4, 1, 6, 7, 2, 3, 8]
values2 = [8, 3, 2, 7, 6, 1, 4, 9, 8, 5]
plt.plot(
    range(1, 11), values1, c="m", lw=1, ls="--", marker=">"
)  # magenta dashed line with triangle markers
plt.plot(
    range(1, 11), values2, c="c", lw=1, ls=":", marker="o"
)  # cyan dotted line with circle markers
plt.show()
```

# Labels, Legends, Annotations

```python
values1 = [5, 8, 9, 4, 1, 6, 7, 2, 3, 8]
values2 = [8, 3, 2, 7, 6, 1, 4, 9, 8, 5]
plt.plot(range(1, 11), values1)
plt.plot(range(1, 11), values2)
plt.xlabel("Roll No")
plt.ylabel("CPI")
plt.annotate(xy=[5, 1], text="Lowest CPI")
plt.annotate(xy=[8, 9], text="Best CPI")
plt.legend(["CX", "CY"], loc=4)
plt.show()
```
